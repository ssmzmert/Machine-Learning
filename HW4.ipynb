```
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
```

```
iris = datasets.load_iris()
x = iris.data
```

```
n_cluster = 1
kmeans = KMeans(n_cluster, n_init=1).fit(x)

print("Labels: ")
print(kmeans.labels_, end='\n\n')

print("Cluster Centers: ")
print(kmeans.cluster_centers_, end='\n\n')

print("Inertia: ")
print(kmeans.inertia_)
```

```
n_cluster = 3
kmeans = KMeans(n_cluster, n_init=1).fit(x)

print("Labels: ")
print(kmeans.labels_, end='\n\n')

print("Cluster Centers: ")
print(kmeans.cluster_centers_, end='\n\n')

print("Inertia: ")
print(kmeans.inertia_)
```

```
n_cluster = 10
kmeans = KMeans(n_cluster, n_init=1).fit(x)

print("Labels: ")
print(kmeans.labels_, end='\n\n')

print("Cluster Centers: ")
print(kmeans.cluster_centers_, end='\n\n')

print("Inertia: ")
print(kmeans.inertia_)
```

```
n_cluster = 50
kmeans = KMeans(n_cluster, n_init=1).fit(x)

print("Labels: ")
print(kmeans.labels_, end='\n\n')

print("Cluster Centers: ")
print(kmeans.cluster_centers_, end='\n\n')

print("Inertia: ")
print(kmeans.inertia_)
```

```
n_cluster = 149
kmeans = KMeans(n_cluster, n_init=1).fit(x)

print("Labels: ")
print(kmeans.labels_, end='\n\n')

print("Cluster Centers: ")
print(kmeans.cluster_centers_, end='\n\n')

print("Inertia: ")
print(kmeans.inertia_)
```

```
#This part is when n_init is equal to 10
n_clusters = [1,3,10,50,149]

for i in n_clusters:
  kmeans = KMeans(n_clusters=i, n_init=10).fit(x)
  print("\nCluster value: ", i)
  print("\nLabels: ")
  print(kmeans.labels_)
  print("\nCluster Centers: ")
  print(kmeans.cluster_centers_)
  print("\nInertia: ")
  print(kmeans.inertia_)
```

As we can see from the inertia values n_cluster=1 has the highest value which means that the error values between nodes to the cluster is the highest. When n_cluster=149 we see that the inertia value becomes 0 with no error rate. However, this might not be useful and efficient in real life because it becomes more costly as the number of clusters increase.
