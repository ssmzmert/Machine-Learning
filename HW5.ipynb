```
from keras.datasets import mnist
from sklearn.neural_network import MLPClassifier

(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Flatten the images
image_vector_size = 28*28
x_train = x_train.reshape(x_train.shape[0], image_vector_size)
x_test = x_test.reshape(x_test.shape[0], image_vector_size)
```

```
# run 1 : size 300 and 100 Neurons
classifier = MLPClassifier(hidden_layer_sizes=(300,100), solver='sgd', activation='relu').fit(x_train, y_train)
classifier_t = MLPClassifier(hidden_layer_sizes=(300,100), solver='sgd', activation='tanh').fit(x_train, y_train)

print("Relu score: ",classifier.score(x_test, y_test))
print("Tanh sccore: ",classifier_t.score(x_test, y_test))
```

```
#run 2: 100 and 50 Neurons
classifier_2 = MLPClassifier(hidden_layer_sizes=(100,50), solver='sgd', activation='relu').fit(x_train, y_train)
classifier_2_t = MLPClassifier(hidden_layer_sizes=(100,50), solver='sgd', activation='tanh').fit(x_train, y_train)

print("Relu score: ",classifier_2.score(x_test, y_test))
print("Tanh sccore: ",classifier_2_t.score(x_test, y_test))
```
