1)Use Pandas to read the iris dataset from
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

```
import numpy as np
import pandas as pd

url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris_data = pd.read_csv(url, names=["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Output"])
```

```
iris_data.columns
iris_data
```

```
from sklearn import datasets
iris = datasets.load_iris()

data, labels = iris.data, iris.target

data.shape, labels.shape
```

**2)Apply kNN algorithm with k=1 to 10 ( you can use a for loop for this purpose ) . Use %75 to %25 ratio for training/testing split and "Uniform" weights**

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.25, random_state=5) # X_train_v -> 75%, X_test -> 25%
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```
from sklearn.preprocessing import MinMaxScaler

# Apply feature scaling
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```
iris.feature_names
```

a) Run the full code three times with same parameters for each k, explain why you may be getting different results on each run with same parameters.

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report

kVals = range(1, 11, 1)
accuracies = []

for k in kVals:
  # Train the k-Nearest Neighbor classifier with the current value of k
 model = KNeighborsClassifier(n_neighbors=k)
 model.fit(X_train, y_train)

  # Evaluate the model on validation set 
 val_predictions = model.predict(X_test)


 print("For k = {:d}".format(k))
 print("Test accuracy = {:.2f}%".format(accuracy_score(y_test , val_predictions) *100 ))

```

b) Decide which k performs better with this dataset by looking confusion matrix and F1 scores

k = 6, 7, 8 performs better with 94.74% accuracy. Since 6 comes first, we use 6 as k.

```
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

model = KNeighborsClassifier(n_neighbors=kVals[6])
model.fit(X_train, y_train)

y_preds = model.predict(X_test)

print("Best k = 6")
print("Test accuracy = {:.2f}%".format(accuracy_score(y_test, y_preds) * 100))
```

```
print(classification_report(y_test, y_preds))
```

```
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_preds))
```

```
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt

plot_confusion_matrix(model, X_test, y_test, cmap = plt.cm.Blues)
plt.show()
```
